```python
df = pd.read_csv('fifa21_train.csv')
df.head()
df.tail()
```

```python
def preprocess(df):
    cols = []
    for column in df.columns:
        cols.append(column.lower())
    df.columns = cols
    df.dropna(axis=0,how='all',inplace=True)
    df.dropna(axis=1,how='all',inplace=True)
    return df
df = preprocess(pd.read_csv('fifa21_train.csv'))
df.head()
```
